```python
import numpy as np

def my_fit(X, y):
    N, D = X.shape
    learning_rate = 2
    num_iterations = 20
    S= 512
    w = np.linalg.lstsq(X, y, rcond=None)[0]

    for t in range(num_iterations):
        # Compute the gradient of the loss function
        gradient = np.dot(X.T, (np.dot(X, w) - y)) / N

        # Update w using gradient descent
        z = w - learning_rate * gradient

        # Apply sparsity constraint using HT(⋅, S)
        indices = np.argpartition(np.abs(z), -S)[-S:]
        # Solve S-dimensional least squares problem
        active_cdus = np.isin(np.arange(D), indices)
        u = np.linalg.lstsq(X[:, active_cdus], y, rcond=None)[0]

        w = np.zeros(D)
        w[active_cdus] = u

    return w


# Assuming you have the training challenges and responses data
X= np.genfromtxt('train_challenges.dat', delimiter = ' ')
y=np.genfromtxt('train_responses.dat', delimiter = ' ')

n= X.shape[0]
n_val = int(0.2* n)
train_X = X[:-n_val]
train_y = y[:-n_val]
val_X = X[-n_val:]
val_y = y[-n_val:]

# Learn the linear model
w = my_fit(train_X, train_y)

def predict(X,w):
 return np.dot(X,w)

predicted_y=predict(val_X,w)
error= predicted_y-val_y
distance = np.linalg.norm(predicted_y - val_y)
mean_absolute_error = np.mean(np.abs(predicted_y - val_y))

print(distance)
print(mean_absolute_error)
print(predicted_y)
print(val_y)
print(np.count_nonzero(w))
```
